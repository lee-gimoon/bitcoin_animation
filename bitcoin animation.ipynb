```python
import pandas as pd
import plotly.graph_objects as go
df = pd.read_csv('C:\\Users\\yy123\\OneDrive\\바탕 화면/btc123.csv')
df = df.set_index('날짜')
df = df.sort_index(ascending=True)
df = df.reset_index()

fig = go.Figure(
                data=go.Scatter(x=df['날짜'], y=df['종가'],mode='lines'),
                layout=go.Layout(updatemenus=[dict(buttons = [dict(
                                               args = [None, {"frame": {"duration": 100}, 
                                                "transition": {"duration":80}}  ],               
                                                               
                                                              
                                               label = "Play",
                                               method = "animate")])]
                 ),
                frames=[go.Frame(data=[go.Scatter(x=df['날짜'][:frame],y=df['종가'][:frame])]
                    
                   )for frame in range(1,1570)]
)
import dash
import dash_core_components as dcc
import dash_html_components as html

app = dash.Dash()
app.layout = html.Div([
    dcc.Graph(figure=fig)
])

app.run_server(debug=True, use_reloader=False)


fig.show()
```
